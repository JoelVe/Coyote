## Datos:
$$b=3[m]$$
$$\mu=0.3$$
$$m=2000 [kg]$$


Para la resolución se requiere de un análisis de fuerzas en el cuerpo, según el siguiente DCL. 
Con sus respectivos ejes coordenados. 



Se sabe que:
$$\sum \vec{\tau} = \vec{\frac{\partial H/A}{\partial t}} + \vec{V_{A/O}} \times \vec{p_{O}} $$
Para este caso la piedra rota alrededor de O por lo tanto:
$$\sum \vec{\tau}/O = \vec{\frac{\partial H/O}{\partial t}}=I_{zz} \vec{\ddot{\omega}} $$
Ya que:
$$\vec{V_{A/O}} \times \vec{p_{O}}=0$$
El eje de giro de la piedra no es un eje principal, sin embargo, es un eje paralelo, por lo tanto por el teorema de los ejes paralelos se tiene:
$$I_{zz}/A=I_{zz}/G+md^{2}$$
La inercia para un cubo que gira con respecto a su centroide en z es:
$$I_{zz}/G=\frac{mb^{2}}{6}$$
Por el teorema de Steiner se tiene:
$$I_{zz}/A=\frac{mb^{2}}{6}+\frac{mb^{2}}{4}=\frac{5}{12}mb^{2}$$
Adicionalmente se tiene:
$$\sum \vec{\tau}/O =\sin{\theta}*b/2*mg\vec{k}$$
Finalmente tenemos:
$$\sum \vec{\tau}/O =I_{zz} \vec{\ddot{\omega}}$$
$$mg*b/2*\sin{\theta}=\frac{5}{12}mb^{2}\ddot{\theta}$$
$$\ddot{\theta}=\frac{12g}{5b}\sin{\theta}$$
Luego para hallar la velocidad angular tenemos:
$$\int_{0}^{\dot{\theta}}\dot{\theta} d\dot{\theta}=\int_{0}^{\theta}\ddot{\theta}d\theta$$
$$\dot{\theta}^{2}=\frac{12g}{5b} \int_{0}^{\theta}\sin{\theta}d\theta$$
$$\dot{\theta}^{2}=\frac{12g}{5b}(1-\cos{\theta})$$
Si $g=9.81$ y $b=3 [m]$
$$\dot{\theta}^{2}=7.848 (1-\cos{\theta})$$

Para el análisis de fuerzas tenemos:
$$\sum F_{n}=m a_{n}$$
Se sabe que:
$$\dot{\theta}^{2}r=a_{n}$$
Entonces:
$$mg \cos{\theta}-N=m \frac{b}{2}(\frac{12g}{5b}(1-\cos{\theta})$$
Con identidades trigonométricas se tiene:
$$N=\frac{mg}{5}(11\cos{\theta}-6)=3924(11\cos{\theta}-6$$
$$\sum F_{t}=m a_{t}$$
$$mg \sin{\theta}-fr=m\frac{b}{2}\frac{6}{5}\frac{g}{b}\sin{\theta}$$
$$fr=\frac{2}{5}mg\sin{\theta}=7848\sin{\theta}$$
$$fr=\mu N=0.3 N $$
De donde se tiene:
$$\frac{2}{5}mg\sin{\theta}=0.3 \cdot\frac{mg}{5}(11\cos{\theta}-6)$$
$$=7848\sin{\theta}=0.3 \cdot 3924(11\cos{\theta}-6)$$
$$\theta = 30.97=31^{\circ}$$
Que se puede interpretar como el ángulo que empezará a deslizar. 
Para calcular el ángulo en el que la piedra deja de tener contacto con el borde se deberá cumplir
$$N=0$$
$$N=\frac{mg}{5}(11\cos{\theta}-6)=0$$
$$\theta=56.9^{\circ}$$


Finalmenta se puede decir que lo ocurrido en el video no es posible puesto que las condiciones se prestan para entender que la piedra rotará hacia donde su masa se concentre en su mayoría, que con lo observado esta ya se encontraba vecinda prácticamente. 


```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go
#Resolución de gráfico
theta=np.arange(0,56.7,0.2) 
vtheta = np.array(th) 
w2=(12/5)*9.81/3*(1-np.cos(theta/180*np.pi))
#plot
fig=px.line(x=theta,y=w2,labels={'x':'Ángulo de inclinación de la piedra con respecto al suelo [°]','y':'Velocidad angular [rad/s]'},title="Velocidad angular  vs ángulo de giro")
fig.show()
```
